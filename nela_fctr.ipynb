```python
import pandas as pd
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from nela_features.nela_features import NELAFeatureExtractor
```

```python
nela = NELAFeatureExtractor()
DATA = "snopes"
```

```python
if DATA.startswith("fctr"):
    DEFAULT_SYSTEM_PROMPT = """Is the following statement in Turkish true or false? Answer as either \"doğru\" or \"yanlış\" """.strip()
    
    df = pd.read_csv("data/fctr.csv", sep="\t", encoding='utf-8')
    df = df[['claim', 'summary', 'label']]
    df = df.dropna()
    df.columns = ['claim', 'evidence', 'label']
    
    if DATA == "fctr500":
        df3 = df[df.label == "yanlış"].sample(n=297, random_state=42)
    elif DATA == "fctr1000":
        df3 = df[df.label == "yanlış"].sample(n=797, random_state=42)

    df4 = df[df.label == "doğru"]
    df = pd.concat([df3, df4])
    df = df.sample(frac=1, random_state=42).reset_index(drop=True)
    #df.replace({'label': {"doğru": 0, "yanlış": 1}}, inplace=True)

elif DATA == "snopes":
    DEFAULT_SYSTEM_PROMPT = """Is the following statement \"true\" or \"false\" ?""".strip()
    
    df = pd.read_csv("snopes/snopes.csv", sep="\t", encoding="utf-8")
    df = df[["claim", "evidence", "label"]]
    b = ["true", "false"]
    df= df[df['label'].isin(b)]
    df.dropna(inplace=True)
    #df.replace({'label': {"true": 1, "false": 0}}, inplace=True)
    
df.head()
```

FCTR1000

```python
feature_vector, feature_names = nela.extract_all(df.iloc[0]["claim"])

data = {}
labels = []
for i in range(len(feature_names)):
    data[feature_names[i]] = []
    
for idx, row in df.iterrows():
    labels.append(row["label"])

    feature_vector, feature_names = nela.extract_all(row["claim"])
    for i in range(len(feature_names)):
        data[feature_names[i]].append(feature_vector[i])


for key in data.keys():
    result = pairwise_tukeyhsd(data[key], labels)
    if result.reject[0]:
        print(key)
        print(result)
        print("**********************************")  
```
